```python

import numpy as np
import h5py
import os
import numpy as np
import matplotlib.pyplot as plt

# import Dataset

from google.colab import drive
drive.mount('/content/drive')

################################################################

# Loading multiple polygons algorithm

################################################################

def load_frame_keys(filename):
    data = h5py.File(filename, "r")
    root = data["/mass"]
    return root.keys()

def load_frame_data(filename, key):
    data = h5py.File(filename, "r")

    root = data["/area"]
    area = np.array(root[key])

    root = data["/mass"]
    mass = np.array(root[key])

    root = data["/u"]
    u = np.array(root[key])

    root = data["/v"]
    v = np.array(root[key])

    vel = np.stack((u, v), axis=1)

    root = data["/coords"]
    coords = [np.array([np.array([root[eee] for eee in root[ee]]) for ee in root[e]]) for e in root[key]]

    polys = []
    for p in coords:

        for pp in p:
            x = [pt[0] for pt in pp]
            y = [pt[1] for pt in pp]
            polys.append(np.array([x, y]))

    root = data["/stress"]
    stress = np.array([np.array(data[e]) for e in root[key]])

    root = data["/centroid"]
    centroid = np.array([np.array(data[e]) for e in root[key]])

    return polys, centroid, vel, stress, mass

datafile_DEM = "/content/drive/MyDrive/Colab Notebooks/floes.jld2"  # data file path
frame_keys = range(0, 1000, 50)

for fkey in frame_keys:
    polys, centroid, vel, stress, mass = load_frame_data(datafile_DEM, str(fkey))
    # print(polys) (debugging)


#debugging
# Get available keys
# if os.path.isfile(datafile_DEM):
    #available_keys = load_frame_keys(datafile_DEM)
    #print(f"Available keys: {available_keys}")

    #frame_keys = [str(fkey) for fkey in range(0, 20000, 50) if str(fkey) in available_keys]
    #print(f"Valid frame keys: {frame_keys}")

    # Process data for valid keys
    # for fkey in frame_keys:
        # polys, centroid, vel, stress, mass = load_frame_data(datafile_DEM, fkey)


#########################################################

#drawing smoothened shapes (smoothened version of the ice particles)

###########################################################

import numpy as np
import matplotlib.pyplot as plt

# Function to calculate distances between points
def calculate_distances(points):
    distances = []
    for i in range(len(points)):
        p1 = points[i]
        p2 = points[(i + 1) % len(points)]
        distance = np.linalg.norm(np.array(p2) - np.array(p1))  # sqrt(p2 - p1)
        distances.append(distance)
    return distances

# Function to get fractional distances of each edge on 2pi range
def calculate_t_values(distances, total_perimeter):
    t_values = [0]
    cumulative_distance = 0
    for d in distances:
        cumulative_distance += d
        t_values.append((cumulative_distance / total_perimeter) * 2 * np.pi)
    return t_values

# Function to create the piecewise function
def piecewise_function(t, points, t_values):
    n = len(points)
    for i in range(n):
        if t_values[i] <= t < t_values[i + 1]:
            a = points[i]
            b = points[(i + 1) % n]
            T1 = t_values[i]
            T2 = t_values[i + 1]
            x_t = a[0] + ((t - T1) / (T2 - T1)) * (b[0] - a[0])
            y_t = a[1] + ((t - T1) / (T2 - T1)) * (b[1] - a[1])
            return (x_t, y_t)
    return points[0]  # In case t == 2*pi, return the first point

# Function to transform and smooth the shape
def transform_shape(points, num_points=100):
    distances = calculate_distances(points)
    total_perimeter = sum(distances)
    t_values = calculate_t_values(distances, total_perimeter)

    t_samples = np.linspace(0, 2 * np.pi, num_points)
    transformed_points = [piecewise_function(t, points, t_values) for t in t_samples]
    x = np.array([p[0] for p in transformed_points])
    y = np.array([p[1] for p in transformed_points])

    return x, y

# Function to plot the original and filtered polygons
def plot_polygons_and_filtered_fft(polys, cutoff, num_points=1000):
    plt.figure(figsize=(18, 5))

    for idx, points in enumerate(polys):
        x, y = transform_shape(points.T, num_points)

        # Plot the original polygon
        plt.subplot(1, 3, 1)
        plt.plot(x, y, 'b-')
        plt.fill(x, y, 'skyblue', alpha=0.3)
        plt.title('Original Polygons in 2D')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.axis('equal')
        plt.grid(True)

        # Apply FFT
        z = x + 1j * y
        fft_z = np.fft.fft(z)

        # Plot FFT
        plt.subplot(1, 3, 2)
        plt.plot(np.abs(fft_z), 'r-')
        plt.title('FFT of the Polygons Boundary')
        plt.xlabel('Frequency')
        plt.ylabel('Magnitude')
        plt.grid(True)

        # Zero out high frequencies
        fft_z_filtered = fft_z.copy()
        fft_z_filtered[cutoff:-cutoff] = 0

        # Apply inverse FFT
        z_filtered = np.fft.ifft(fft_z_filtered)
        x_filtered = np.real(z_filtered)
        y_filtered = np.imag(z_filtered)

        # Plot the filtered polygon
        plt.subplot(1, 3, 3)
        plt.plot(x_filtered, y_filtered, 'g-')
        plt.fill(x_filtered, y_filtered, 'lightgreen', alpha=0.3)
        plt.title('Filtered Polygons in 2D')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.axis('equal')
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# Example usage with polys array
cutoff = 5  # Adjust the cutoff frequency as needed
plot_polygons_and_filtered_fft(polys, cutoff)
```
